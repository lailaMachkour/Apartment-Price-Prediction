```python
import pandas as pd

# Initialize a list to store data for all apartments
apartment_data = []

# Define the titles to be used as columns in the DataFrame
titles = []

num_pages = 125

for page in range(101,num_pages + 1):
    url = f"https://www.avito.ma/fr/maroc/appartements?o={page}"  # Change the URL to navigate through pages
    driver.get(url)
    src = driver.page_source
    soup = BeautifulSoup(src, 'html.parser')

    # Find all apartment listings on the page
    listings = soup.find_all('div', class_='sc-1nre5ec-1 crKvIr listing')

    for listing in listings:
        ds = listing.find_all('a', class_='sc-1jge648-0')

        if ds:
            # Extract information from the listing
            for i in range(len(ds)):
                apartment_dict = {}  # Dictionary to store data for this apartment

                nom = ds[i].find('p', {'class': 'sc-1x0vz2r-0 czqClV'}).get_text().replace('\n', '')
                prix = ds[i].find('div', {'class': 'sc-b57yxx-0 ckugTG'}).find('div', {'class': 'sc-b57yxx-4 dRjnHr'}).get_text().strip().replace('DH', '')
                ville = ds[i].find('div', {'class': 'sc-b57yxx-8 cnMHvz'}).find('p').get_text()
                dat = ds[i].find('div', {'class': 'sc-1wnmz4-2 cA-DLBz'}).find('p').get_text()
 
                apartment_dict['Nom'] = nom
                apartment_dict['Prix'] = prix
                apartment_dict['Ville'] = ville
                apartment_dict['Date'] = dat

                url2 = ds[i].get('href')
                driver.get(url2)
                src2 = driver.page_source
                soup2 = BeautifulSoup(src2, 'html.parser')

                # Extract information from each apartment page
                location_element = soup2.find('div', class_='sc-1g3sn3w-7 bNWHpB')
                if location_element:
                    location = location_element.find_all('span')[0].get_text()
                

                ds2 = soup2.find('div', class_='sc-1g3sn3w-3 fHaZgB')

                # Collect titles and info for this apartment
                title_data = []
                info_data = []

                for j in range(len(ds2)+1):
                    if j < len(ds2.find_all('div', {'class': 'sc-6p5md9-1'})):
                        title_element = ds2.find_all('div', {'class': 'sc-6p5md9-1'})[j].find('title')
                        if title_element:
                            title = title_element.get_text()[:-5]
    
                        info = ds2.find_all('div', {'class': 'sc-6p5md9-1'})[j].find('span').get_text()
                        title_data.append(title)
                        info_data.append(info)
                    else:
    # Handle the case where j is out of range, e.g., add default values or skip this iteration
                        continue  # Skip this iteration or add default values

                # Create a dictionary with empty values for all titles
                apartment_columns = {title: '' for title in titles}

                # Fill in the values for this apartment
                for t, i in zip(title_data, info_data):
                    apartment_columns[t] = i

                # Add the location information to the dictionary
                apartment_dict['Location'] = location
                
                combined_apartment = apartment_columns.copy()  # Create a copy of apartment_columns
                combined_apartment.update(apartment_dict)
                
                # Append the apartment data dictionary to the list
                apartment_data.append(combined_apartment)


```
